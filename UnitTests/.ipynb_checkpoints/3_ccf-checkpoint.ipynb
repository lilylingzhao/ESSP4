```python
# Translate Fit Option Names to fit_rv inputs
model_dict = {'gaussian':ccf_utils.Model_Gaussian,
              'hermite':ccf_utils.Model_H0_CCF0,
              'quadratic':ccf_utils.ccfQuad}
def getFitParams(fit_name=None):
    fitParams = {
        'vrange':1e6,
        'sigma_v':1e5,
        'fit_opts':ccf_utils.fit_opts,
        'opt_method':'lm',
        'model':ccf_utils.Model_Gaussian,
        'rv_guess':0,
    }
    if fit_name is None:
        return fitParams
    
    mods = fit_name.split('-')
    for m in mods:
        try:
            key, val = m.split('_')
            if key=='rv0':
                key = 'rv_guess'
            fitParams[key] = float(val)
        except ValueError:
            fitParams['model'] = model_dict[m]
    return fitParams
```

```python
# Modded version of fit_rv from expres_pipeline/expres/rv/ccf/fit_rv
def fit_rv(v_grid, ccf, e_ccf, vrange=1e6, sigma_v=1e5, fit_opts=ccf_utils.fit_opts,
           invert=False, opt_method='lm', model=ccf_utils.Model_Gaussian, rv_guess=None):
    """Find a radial velocity by way of fitting a Gaussian to the CCF.

    Sometimes, this returns an OptimizeWarning (unable to determine covariances).
    Usually, this is because the default method of determining the correct
    ε to use for finite-difference evaluation of the Jacobian is failing.
    Either supply an analytic Jacobian, or supply a guess for ε
    via the `epsfcn` kwarg in fit_opts. e.g. epsfcn=.1 works pretty well
    for LFC diagnostic velocities.

    Parameters
    ----------
    v_grid : ndarray
        Array of velocity gridpoints at which CCF was evaluated
    ccf : ndarray
        The cross-correlation function to fit
    e_ccf : ndarray
        Error estimate of the CCF
    makeplots : bool
        Make diagnostic plots (default False)
    vrange : number
        velocity range to use in fit (to exclude noisy continuum)
    sigma_v : number
        Initial guess for velocity width of the CCF.
    fit_opts : dict
        Options for curve_fit
    invert : bool
        Whether or not the CCF was inverted
    opt_method : str
        Least squares method to pass into curve_fit
    ret_chi2 : bool
        Return the chi-squared of the fit along with the results

    Returns
    -------
    rv : float
        The fit radial velocity
    rv_err : float
        The error of the fit radial velocity
    chi2 : float (optional)
        The chi-squared of the RV fit
    """

    # Set the guesses for the fit parameters
    offset = np.max(ccf) if not invert else np.min(ccf)
    amp = (np.min(ccf) - np.max(ccf)) * (-1 if invert else 1)
    if rv_guess is None:
        rv_guess = v_grid[np.argmin(ccf) if not invert else np.argmax(ccf)]
    p0 = [amp, rv_guess, sigma_v, offset]

    # Mask velocities outside vrange
    vmask = np.ones_like(v_grid, dtype=bool)
    if vrange is not None:
        vmask = np.abs(v_grid - rv_guess) < vrange
    vmask &= np.isfinite(ccf) & (e_ccf > 0.0)

    if np.sum(vmask) < 1:
        return model, np.nan, np.nan
    
    try:
        popt, pcov = curve_fit(model, v_grid[vmask],
                               ccf[vmask], p0=p0, sigma=e_ccf[vmask],
                               absolute_sigma=True, method=opt_method, **fit_opts)
    except RuntimeError:
        rv, rv_err = np.nan, np.nan
        popt = np.nan
        pcov = np.nan
    else:
        if np.isnan(pcov[1, 1]) or pcov[1, 1] <= 0.0:
            rv, rv_err = np.nan, np.nan
        else:
            rv = popt[1]
            rv_err = np.sqrt(pcov[1, 1])
    
    return model, popt, pcov
```

## Check All Order-by-Order RVs
Plot all the RVs against each other.  Do it order-by-order too.

```python
def compareRvPlot(test,dint,inst,fig=None,axes=None):
    icol = np.arange(len(instruments))[np.array(instruments)==inst][0]
    df_full = df_dict[inst]
    dint_mask = df_full['Time [MJD]'].astype(int)==dint
    df = df_full.copy()[dint_mask]
    time, rvel, errs = df.loc[:,['Time [MJD]','RV [m/s]','RV Err. [m/s]']].to_numpy().T
    rvel -= np.nanmedian(rvel)

    # Collect OBO RVs
    orders = None
    for ifile,file in enumerate(df['File Name']):
        ccf_file = os.path.join(ccf_dir,test,inst,file)
        ccf_dict = readCCF(ccf_file,standard=True)
        if orders is None:
            orders = ccf_dict['orders'].copy()
            obo_rvs = np.full((np.sum(dint_mask),len(orders)),np.nan)
        obo_rvs[ifile,:] = ccf_dict['obo_rv'].copy()
    for i,o in enumerate(orders):
        obo_rvs[:,i] -= np.nanmedian(obo_rvs[:,i])

    # Subplots on Subplots
    num_ord = len(orders)
    num_col = 8
    num_row = num_ord//num_col + (1 if num_ord%num_col==0 else 0)

    fig, axes = plt.subplots(num_row,num_col,figsize=(num_col*3,num_row*1.5))
    axes = np.concatenate(axes)
    for i,o in enumerate(orders[:-1]):
        ax = axes[i]
        ax.set_xticks([])
        ax.set_yticks([])
        ax.plot(time,rvel,'k.')
        ax.plot(time,obo_rvs[:,i],'.',color=inst_cols[icol])
        ax.set_ylim(-10,10)

        ax.text(0.01,0.98,f'{o}',va='top',ha='left',transform=ax.transAxes,fontsize=14)
    #fig.tight_layout()
    return fig, axes
```

```python
def ccfResid(file_name): # name of read in file, but I guess eventually will be the standard file format
    drp_file = getDrpCcfFile(file_name)
    
    if not os.path.isfile(drp_file):
        #print(f'Cannot Match to DRP File: {drp_file}')
        return
    
    ccf_dict = readCCF(file_name,standard=True)
    drp_dict = readCCF(drp_file,standard=False)
    
    # Normalize CCFs
    ccf_norm = ccf_dict['ccf']/(np.max(ccf_dict['ccf'])-np.min(ccf_dict['ccf']))
    ccf_norm -= np.nanmin(ccf_norm)
    drp_norm = drp_dict['ccf']/(np.max(drp_dict['ccf'])-np.min(drp_dict['ccf']))
    drp_norm -= np.nanmin(drp_norm)
    
    ccf_dict['ccf_norm'] = ccf_norm.copy()
    drp_dict['ccf_norm'] = drp_norm.copy()
    
    # Interpolate
    v_intp = np.arange(-20,20,0.2)
    ccf_intp = np.interp(v_intp,ccf_dict['v_grid'],ccf_norm,
                         left=np.nan,right=np.nan)
    drp_intp = np.interp(v_intp,drp_dict['v_grid'],drp_norm,
                         left=np.nan,right=np.nan)
    
    # Align and Interpolate
    ccf_algn = np.interp(v_intp,ccf_dict['v_grid']-ccf_dict['rv']/1000,ccf_norm,
                         left=np.nan,right=np.nan)
    drp_algn = np.interp(v_intp,drp_dict['v_grid']-drp_dict['rv']/1000,drp_norm,
                         left=np.nan,right=np.nan)
    
    ccf_dict['v_intp'] = v_intp.copy()
    ccf_dict['r_intp'] = ccf_intp-drp_intp
    ccf_dict['r_algn'] = ccf_algn-drp_algn
    
    return ccf_dict, drp_dict
```

## Test Residuals to Gaussian Fits
What is the best fit to the CCF?  Any suspicious structure in these residuals?

```python
def fitResidPlot(file_name,resid=True):
    inst = fileName2Inst(file_name)
    fig, axes = plt.subplots(len(rvfit_list)+1,len(test_list),figsize=(len(test_list)*3,(len(rvfit_list)+1)*3))
    for itest, test in enumerate(test_list):
        # Load in CCF to be fit
        ccf_file = os.path.join(ccf_dir,test,inst,file_name)
        if inst in ['harps','harpsn']:
            file_name = os.path.join(solar_dir,instrument_nickname2Fullname(inst),file_name)
            ccf_file = os.path.join(solar_dir,'UniformCCFs',test,inst,
                                    os.path.basename(getHarpsBlazeFile(file_name)))
        ccf_dict, drp_dict = ccfResid(ccf_file)

        # Plot Residual of this CCF vs. DRP
        ax = axes[0,itest]
        ax.plot(ccf_dict['v_intp'],ccf_dict['r_intp'],'C0')
        ax.plot(ccf_dict['v_intp'],ccf_dict['r_algn'],'C1')

        ccf_x, ccf_y, ccf_e = ccf_dict['v_grid'], ccf_dict['ccf'], ccf_dict['e_ccf']
        drp_x, drp_y, drp_e = drp_dict['v_grid'], drp_dict['ccf'], drp_dict['e_ccf']
        ccf_x *= 1e5
        drp_x *= 1e5
        if np.sum(np.isfinite(drp_e))==0:
            drp_e = np.sqrt(drp_y)
        # Plot residual of this CCF vs. various Gaussian Fits
        for ifit,rvFit in enumerate(rvfit_list):
            ax = axes[ifit+1,itest]
            ax.set_title(f'{test}\n{rvFit}')

            # DRP Fit
            model, popt, pcov = fit_rv(drp_x, drp_y, drp_e,**getFitParams(rvFit))
            if np.sum(np.isnan(popt))==0:
                if resid:
                    ax.plot(drp_x/1e5,drp_y-model(drp_x,*popt),'k')
                else:
                    ax.plot(drp_x/1e5,drp_y,'C0-')
                    ax.plot(drp_x/1e5,model(drp_x,*popt),'C2-')
                ax.axvline(popt[1]/1e5,color=f'k',label=round(popt[1],3))

            # Standard CCF Fit
            model, popt, pcov = fit_rv(ccf_x, ccf_y, ccf_e,**getFitParams(rvFit))
            if np.sum(np.isnan(popt))==0:
                if resid:
                    ax.plot(ccf_x/1e5,ccf_y-model(ccf_x,*popt),'--',color=f'C{ifit}')
                else:
                    ax2 = ax.twinx()
                    ax2.plot(ccf_x/1e5,ccf_y,'-',color=f'C1')
                    ax2.plot(ccf_x/1e5,model(ccf_x,*popt),'C3')
                ax.axvline(popt[1]/1e5,color=f'C{ifit}',label=round(popt[1],3))
        
        ax.set_xlim(-20,20)
        #ax.legend(loc=2)
    fig.tight_layout()
```
